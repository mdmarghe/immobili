```python
y=df[target]
X=df[labels]
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train,X_test,y_train,y_test=train_test_split(X,y,
                                               train_size=0.2,
                                               random_state=667)
```

```python
y_train.size
```

```python
y_test.size
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
import joblib

## to save a model
joblib.dump(model,'regression_test.pkl')
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred_test=model.predict(X_train)
```

```python
res=pd.DataFrame(list(zip(y_pred,y_test)),columns=['prediction','real'])
```

```python
res
```

```python
res['errore']=res.prediction-res.real
```

```python
res
```

```python
res.errore.mean()
```

```python
plt.figure(figsize=(15,10))
x=np.linspace(0,y_pred.shape[0],y_pred.shape[0])
plt.plot(x,y_pred,'b-')
plt.plot(x,y_pred,'y-')
plt.plot(x,res.errore,'r--')
plt.legend(labels=['prediction','real','errore'])
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2score=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=mean_squared_error(y_test,y_pred, squared=False)
    
print('R2 score:',r2score)
print('MAE:' ,mae)
print('MSE:',mse)
print('RMSE:',rmse)
```
